```python
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Data Exploration**

```python


def visualize_two_classes(normal_dir, pneumonia_dir, samples_per_class=6):
    plt.figure(figsize=(15, 6))

    # NORMAL images
    normal_images = os.listdir(normal_dir)[:samples_per_class]
    for i, img_name in enumerate(normal_images):
        img_path = os.path.join(normal_dir, img_name)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(2, samples_per_class, i+1)
        plt.imshow(img)
        plt.title("NORMAL")
        plt.axis('off')

    # PNEUMONIA images
    pneumonia_images = os.listdir(pneumonia_dir)[:samples_per_class]
    for i, img_name in enumerate(pneumonia_images):
        img_path = os.path.join(pneumonia_dir, img_name)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(2, samples_per_class, samples_per_class + i + 1)
        plt.imshow(img)
        plt.title("PNEUMONIA")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Paths to dataset folders
normal_path = 'chest_xray/chest_xray/train/NORMAL'
pneumonia_path = 'chest_xray/chest_xray/train/PNEUMONIA'


visualize_two_classes(normal_path, pneumonia_path)
```

## **Data Proccessing**

```python


IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,          # normalize pixels
    rotation_range=20,       # augmentation
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

# Load train data
train_generator = train_datagen.flow_from_directory(
    'chest_xray/chest_xray/train',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

# Load validation data
val_generator = val_test_datagen.flow_from_directory(
    'chest_xray/chest_xray/val',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary'
)

# Load test data
test_generator = val_test_datagen.flow_from_directory(
    'chest_xray/chest_xray/test',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False
)

```

** MobileNetV2-specific preprocessing**

```python
# Load and preprocess images for MobileNetV2


def load_and_preprocess_images_mobilenet(directory):
    images = []
    labels = []
    classes = ['NORMAL', 'PNEUMONIA']

    for label, class_name in enumerate(classes):
        class_dir = os.path.join(directory, class_name)
        for img_name in os.listdir(class_dir):
            img_path = os.path.join(class_dir, img_name)
            img = cv2.imread(img_path)
            if img is None:
                continue
            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = preprocess_input(img)
            images.append(img)
            labels.append(label)
    return np.array(images), np.array(labels)

train_images, train_labels = load_and_preprocess_images_mobilenet('chest_xray/chest_xray/train')
val_images, val_labels = load_and_preprocess_images_mobilenet('chest_xray/chest_xray/val')
test_images, test_labels = load_and_preprocess_images_mobilenet('chest_xray/chest_xray/test')

print("Train:", train_images.shape, train_labels.shape)
print("Val:", val_images.shape, val_labels.shape)
print("Test:", test_images.shape, test_labels.shape)
```

```python
# Print class indices
print("Class indices:", train_generator.class_indices)

# Number of samples per class
for cls, idx in train_generator.class_indices.items():
    count = sum(train_generator.labels == idx)
    print(f"{cls}: {count} images")

# Total samples
print("Total training images:", train_generator.n)
```

```python

sns.countplot(x=train_generator.labels)
plt.xticks([0,1], list(train_generator.class_indices.keys()))
plt.title("Class Distribution")
plt.show()
```

The chest X-ray training dataset contains a total of 5216 images divided into two classes:

**NORMAL**: 1341 images

**PNEUMONIA**: 3875 images

This shows that the dataset is **imbalanced**, with the PNEUMONIA class having nearly three times more samples than the NORMAL class.


The imbalance may cause a model trained on this data to bias predictions toward the majority class (PNEUMONIA).



## **Model Building**

### **CNN**

```python

input_shape = (IMG_SIZE, IMG_SIZE, 3)

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', padding='same', input_shape=input_shape),
    layers.Conv2D(32, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu', padding='same'),
    layers.Conv2D(64, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), activation='relu', padding='same'),
    layers.Conv2D(128, (3,3), activation='relu', padding='same'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(1, activation='sigmoid')  # Binary classification output
])
```

```python
model.summary()
```

```python



history = model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)


```

```python

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

**Plot training history**

```python
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

```

**Classification report**

```python

# Evaluate on test set
test_loss, test_acc = model.evaluate(test_images, test_labels, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")
# Predict classes on test set
y_pred_probs = model.predict(test_images)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

print(classification_report(test_labels, y_pred, target_names=['NORMAL', 'PNEUMONIA']))
```

**Confusion Matrix visualization**

```python

cm = confusion_matrix(test_labels, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['NORMAL', 'PNEUMONIA'],
            yticklabels=['NORMAL', 'PNEUMONIA'])
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.title('Confusion Matrix')
plt.show()
```

**Training vs Validation Loss & Accuracy curves**

```python


# Accuracy plot
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy over Epochs')

# Loss plot
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss over Epochs')

plt.show()
```

**Conclusion:**
The model demonstrates a clear imbalance in classification behavior, heavily favoring the NORMAL class while almost entirely neglecting the PNEUMONIA class. Although it achieves perfect recall for NORMAL cases, its recall for pneumonia is extremely low, indicating that it fails to identify the majority of actual pneumonia instances.

### **Apply Class Weights During Training**


In the dataset:

**NORMAL** = 1341 images

**PNEUMONIA** = 3875 images

The model sees many more PNEUMONIA images than NORMAL.

If you train without any adjustment, the model might just predict PNEUMONIA all the time to get higher overall accuracy, because that “works” most of the time.


During training, the loss function averages over all samples.

Mistakes on NORMAL images barely affect the total loss, so the model **“ignores”** them and predicts mostly **PNEUMONIA**.

This leads to poor performance on the minority class (NORMAL).

Metrics like accuracy can be misleading: a model predicting only PNEUMONIA could still get ~74% accuracy (3875 / 5216).



```python
from sklearn.utils import class_weight
import numpy as np

# Compute class weights
class_weights = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_generator.labels),
    y=train_generator.labels
)

# Convert to dictionary for Keras
class_weights = dict(enumerate(class_weights))
print("Class weights:", class_weights)
```

 **How class weights fix this?**

Class weights tell the model:

“Mistakes on this class are more important — penalize them more in the loss function.”

class_weights = {0: 1.94, 1: 0.67}
Class 0 = NORMAL → weight 1.94

Class 1 = PNEUMONIA → weight 0.67

This means:

If the model misclassifies a NORMAL image, the loss is **multiplied** by **1.94**.

If it misclassifies a PNEUMONIA image, the loss is **multiplied** by **0.67**.

The training algorithm “pays more attention” to the minority class, so it tries harder to predict NORMAL correctly.

```python
model_with_weights = tf.keras.models.clone_model(model)
model_with_weights.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_with_weights = model_with_weights.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator,
    class_weight=class_weights
)
```

```python

# Evaluate on test set
test_loss, test_acc = model_with_weights.evaluate(test_images, test_labels, verbose=2)
print(f"Test accuracy: {test_acc*100:.2f}%")
# Predict classes on test set
y_pred_probs = model_with_weights.predict(test_images)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

print(classification_report(test_labels, y_pred, target_names=['NORMAL', 'PNEUMONIA']))
```

### **MobileNetV2**

```python

base_model = MobileNetV2(input_shape=(IMG_SIZE, IMG_SIZE, 3),
                         include_top=False,  # remove last classifier layers
                         weights='imagenet')  # use pretrained weights

base_model.trainable = False  # freeze base model

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),  # reduces spatial dims -> vector
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(1, activation='sigmoid')  # binary classification output
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.summary()





```

**Train model**

```python
# Train model

history = model.fit(train_images, train_labels,
                    epochs=10,
                    batch_size=32,
                    validation_data=(val_images, val_labels))
```

#### **Conclusion**

The MobileNetV2 model achieved a test accuracy of 82.53% with a loss of 0.8520, demonstrating a substantial performance improvement over previous CNN architectures. Class-wise analysis shows a precision of 0.98 and recall of 0.54 for the NORMAL class, indicating high reliability when predicting healthy cases but moderate sensitivity, with nearly half of true NORMAL cases misclassified as PNEUMONIA.

For the PNEUMONIA class, the model achieved precision of 0.78 and recall of 0.99, resulting in an F1-score of 0.88. This near-perfect recall indicates the model’s strong sensitivity to pneumonia patterns, minimizing the false negative rate — a critical requirement in clinical diagnostic contexts. However, the trade-off is an elevated false positive rate for pneumonia, as reflected by the reduced recall for NORMAL.

The macro-averaged metrics (precision: 0.88, recall: 0.77, F1: 0.79) confirm balanced performance across classes, while weighted averages (precision: 0.86, recall: 0.83, F1: 0.81) reflect the class distribution in the dataset. These results suggest that MobileNetV2’s pre-trained feature extraction significantly enhanced representational capacity, improving generalization and class separation. Future optimization should focus on reducing false positives for pneumonia through threshold tuning, calibration, or targeted data augmentation without compromising the high sensitivity achieved for the disease class.
